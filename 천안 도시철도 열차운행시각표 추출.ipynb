```python
import pandas as pd
```

```python
df = pd.read_csv("서울교통공사_서울 도시철도 열차운행시각표_20211229.csv" )
df
```

```python
# 결측치 제거하고 분석 (열차도착시간, 열차출발시간 결측치)
# 역사명: 천안 1호선만 추출

df1 = df.iloc[:, 1:].dropna(subset=['열차도착시간', '열차출발시간']).query("역사명 == ['성환', '직산', '두정', '천안', '봉명', '쌍용']").sort_values(["역사코드", "열차도착시간"])
df1.isnull().sum() 
df1.head(50)
```
